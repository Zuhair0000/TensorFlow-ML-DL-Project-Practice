# Tensorflow

```python
import tensorflow as tf
import numpy as np
import pandas as pd
```

```python
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.fashion_mnist.load_data()
```

```python
X_train.shape, X_train.max()
```

```python
X_train = np.expand_dims(X_train, axis=-1)
X_train.shape
```

```python
train_ds = tf.data.Dataset.from_tensor_slices((X_train, y_train)).shuffle(buffer_size=10000).batch(32).prefetch(tf.data.AUTOTUNE)
test_ds = tf.data.Dataset.from_tensor_slices((X_test, y_test)).batch(32).prefetch(tf.data.AUTOTUNE)
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.RandomFlip('horizontal', input_shape=(28, 28, 1)),

    tf.keras.layers.Rescaling(1./255),

    tf.keras.layers.Conv2D(32, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Flatten(),

    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(10, activation='softmax')
])
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',
              loss='SparseCategoricalCrossentropy',
              metrics=['accuracy'])
```

```python
model.fit(train_ds,
          epochs=5,
          validation_data=test_ds)
```

# PyTorch

```python
import torch
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
```

```python
transform = transforms.ToTensor()
```

```python
train_ds = DataLoader(datasets.FashionMNIST(root='data', train=True, transform=transform, download=True), batch_size=32, shuffle=True)
test_ds = DataLoader(datasets.FashionMNIST(root='data', train=False, transform=transform, download=True), batch_size=32, shuffle=False)
```

```python
images, labels = next(iter(train_ds))
print(f"Image batch shape: {images.shape}")
```

```python
import torch.nn as nn
import torch.nn.functional as F

class FashionCNN(nn.Module):
  def __init__(self):
    super(FashionCNN, self).__init__()
    self.conv1 = nn.Conv2d(in_channels=1, out_channels=32, kernel_size=3)
    self.pool1 = nn.MaxPool2d(kernel_size=2)
    self.fc1 = nn.Linear(in_features=5408, out_features=64)
    self.fc2 = nn.Linear(in_features=64, out_features=10)

  def forward(self, x):
    x = F.relu(self.conv1(x))

    x = self.pool1(x)

    x = torch.flatten(x, 1)

    x = F.relu(self.fc1(x))

    x = self.fc2(x)

    return x
```

```python
model = FashionCNN()
print(model)
```

```python
import torch.optim as optim

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

epochs = 5

for epoch in range(epochs):
  running_loss=0.0

  for images, labels in train_ds:
    optimizer.zero_grad()

    outputs = model(images)

    loss = criterion(outputs, labels)

    loss.backward()

    optimizer.step()

    running_loss += loss.item()

  print(f"Epoch {epoch+1}/{epochs} - Loss: {running_loss/len(train_ds):.4f}")
```

```python
model.eval()

correct = 0
total = 0

with torch.no_grad():
  for images, labels in test_ds:
    outputs = model(images)

    _, predicted = torch.max(outputs, 1)

    total += labels.size(0)

    correct += (predicted==labels).sum().item()
```

```python
accuracy = 100 * correct / total
print(f"PyTorch Test Accuracy: {accuracy:.2f}%")
```
